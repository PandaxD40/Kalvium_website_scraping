```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
url = "https://results.eci.gov.in/PcResultGenJune2024/index.htm"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

data = soup.select("tbody .tr")
results = []

for d in data:
    party_name = d.select_one("td").text
    state_url = d.select_one("td a")["href"]
    state_page_url = "https://results.eci.gov.in/PcResultGenJune2024/" + state_url

    state_response = requests.get(state_page_url)
    state_soup = BeautifulSoup(state_response.text, "html.parser")
    rows = state_soup.select("tbody tr")
    state_results = []

    for row in rows:
        s_no = row.select_one("td:nth-child(1)").text
        state = row.select_one("td:nth-child(2) a").text
        candidate_name = row.select_one("td:nth-child(3)").text
        total_votes = row.select_one("td:nth-child(4)").text
        margin = row.select_one("td:nth-child(5)").text

        result = {
            's_no': s_no,
            'state': state,
            'candidate_name': candidate_name,
            'total_votes': total_votes,
            'margin': margin,
        }

        state_results.append(result)

    state_data = {
        'party_name': party_name,
        'results': state_results
    }

    results.append(state_data)

# Save the data to a JSON file
with open('rows.json', 'w') as file:
    json.dump(results, file)

print("Rows data saved to rows.json file.")
df = pd.read_json('rows.json')
display(df)
df.to_excel('rows.xlsx', index=False)
print("Rows data saved to rows.xlsx file.")
```
